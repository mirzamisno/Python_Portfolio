```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

epl_df= pd.read_csv('C:\\Users\\ADMIN\\Downloads\EPL_20_21.csv')
epl_df.head(5)
```

```python
epl_df.info
```

```python
#To check the data info
epl_df.describe()
```

```python
#To check if theres any null in the data set
epl_df.isna().sum()
```

```python
#To create 2 new columns

epl_df['MinsPMatch'] = (epl_df['Mins']/ epl_df['Matches'].astype(int))
epl_df['GPMatch'] = (epl_df['Goals'] / epl_df['Matches'].astype(int))
epl_df.head(5)
```

```python
#To get the Total Goals Scored in the league

Total_Goals= epl_df['Goals'].sum()
print(Total_Goals)
```

```python
#To get the total penalty goals scored in the league

Total_Pen_Goals = epl_df['Penalty_Goals'].sum()
print(Total_Pen_Goals)
```

```python
#To get the total penalty attempted on goals in the league

Total_Penalty_Attempted= epl_df['Penalty_Attempted'].sum()
print(Total_Penalty_Attempted)

#Penalty not scored
Pl_not_scored= Total_Penalty_Attempted - Total_Pen_Goals
print(Pl_not_scored)
```

```python
# Pie chart for penalty missed versus scored

plt.figure(figsize=(13, 6))
Pl_not_scored= Total_Penalty_Attempted - Total_Pen_Goals
data = [Pl_not_scored, Total_Pen_Goals]
labels = ['Penalties missed', 'Penalties scored']
colors = sns.color_palette("Set2")
explode = [0.1,0]
plt.pie(data, labels = labels, wedgeprops ={'edgecolor': 'black'}, explode=explode, shadow=True, 
        colors= colors, autopct= '%1.1f%%')
plt.title('English Premier League 2020-2021 penalty missed vs scored')
plt.tight_layout()
plt.show()
```

```python
# To find out how many unique position

epl_df['Position'].unique()
```

```python
# Player nationalities

Nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
plt.title('Top 10 number of players by nationality')
plt.ylabel('Number of Players')
plt.xlabel('Nationality')
Nationality.head(10).plot(kind = 'bar', width=0.9,figsize=(8,6), color =  sns.color_palette('magma'))
```

```python
#Club with highest number of players

epl_df['Club'].value_counts().nlargest(5).plot(kind = 'barh', width=0.9,figsize=(6,3), color =  sns.color_palette('viridis'))
plt.title("Highest number of players played in the league")
plt.show()

#Club with highest number of players
epl_df['Club'].value_counts().nsmallest(5).plot(kind = 'barh', width=0.9,figsize=(6,3), color =  sns.color_palette('viridis'))
plt.title("Lowest number of players played in the league")
plt.show()
```

```python
#Players based on age group

under20 = epl_df[epl_df['Age'] <20]
under25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
under30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
above30 = epl_df[epl_df['Age'] > 30]
```

```python
x = np.array([under20['Name'].count(),under25['Name'].count(),under30['Name'].count(),above30['Name'].count()])
labels = ["<20","<=25","<=30",">30"]
colors = sns.color_palette("Set2")
plt.pie(x, labels = labels, wedgeprops ={'edgecolor': 'black'}, shadow=True, 
        colors= colors, autopct= '%1.1f%%')
plt.title('English Premier League 2020-2021 age group',fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Different age group based on clubs
plt.title('Age group less than 20 in EPL clubs', fontsize=14)
under20['Club'].value_counts().plot(kind = 'barh', width=0.7, figsize=(12,6), fontsize= 12,color= sns.color_palette('cubehelix'))
```

```python
#Average player age in the club

plt.figure(figsize=(12,6))
sns.boxplot(x = 'Club', y = 'Age', data = epl_df)
plt.xticks(rotation = 90)
plt.show()
```

```python


num_player= epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending = False)
```

```python
# Total assists from each club

asst_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style = 'whitegrid', color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Assists', data = asst_by_clubs.sort_values(by='Assists'),palette = 'Set2')
ax.set_xlabel('Club',fontsize=20)
ax.set_ylabel('Assists',fontsize=20)
plt.xticks(rotation = 45)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of clubs vs total asssist', fontsize=20)
```

```python
# Top 10 Assist in the league

top10_asst = epl_df[['Name','Club','Assists','Matches']].nlargest(n=10, columns = 'Assists')
top10_asst
```

```python
goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style = 'whitegrid', color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Goals', data = goals_by_clubs.sort_values(by='Goals'),palette = 'rocket')
ax.set_xlabel('Club',fontsize=20)
ax.set_ylabel('Goals',fontsize=20)
plt.xticks(rotation = 45)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of clubs vs total goals', fontsize=20)
```

```python
# Pie chart - Goals with assist and without assist

plt.figure(figsize = (14,7))
assists = epl_df['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals without Assist', 'Goal with Assist']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = color, shadow=True, autopct = '%.0f%%')
plt.title('Goals with assist vs without assist', fontsize=22)
plt.show()
```

```python
# Top 10 players with most yellow cards

epl_yellow_card = epl_df.sort_values(by = 'Yellow_Cards', ascending = False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with the most yellow cards', fontsize=20)
barplot = sns.barplot(x= epl_yellow_card['Name'], y= epl_yellow_card['Yellow_Cards'],
                     label = 'Players', color = 'yellow')
plt.ylabel('Number of yellow cards')
barplot.set_xticklabels(barplot.get_xticklabels(),rotation=45)
```
